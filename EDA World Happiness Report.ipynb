Dataset: https://www.kaggle.com/datasets/unsdsn/world-happiness

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("2019.csv")
```

```python
df.dropna(axis=0,inplace=True)
```

## Data Visualization

```python
plt.figure(figsize=(12,8))
sns.distplot(df["Score"])
plt.title("Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df["GDP per capita"], df["Score"])
plt.title("GDP per Capita vs Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df["Healthy life expectancy"], df["Score"])
plt.title("Healthy Life Expectancy vs Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df["Social support"],df["Score"])
plt.title("Social Support vs Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df["Freedom to make life choices"], df["Score"])
plt.title("Freedom to Make Life Choices vs Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df["Perceptions of corruption"], df["Score"])
plt.title("Perceptions of corruption vs Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
df.plot.hexbin(x="GDP per capita", y="Social support", gridsize=15, sharex=False)
plt.title("GDP per Capita vs Social support", size=15)
plt.show()
```

```python
happiest_10_countries = df.iloc[:10]
plt.figure(figsize=(12,8))
sns.barplot(x=happiest_10_countries["Country or region"], y=df["Score"])
plt.xticks(rotation=30)
plt.title("Scores of Happiest 10 Countries'", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=happiest_10_countries["Country or region"], y=df["GDP per capita"])
plt.xticks(rotation=30)
plt.title("GDP per Capita in Happiest 10 Countries", size=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=happiest_10_countries["Country or region"], y=df["Perceptions of corruption"])
plt.xticks(rotation=30)
plt.title("Perceptions of Corruption in Happiest 10 Countries", size=15)
plt.show()
```
